```python
# 측정하는 cell
%reset
# 
%cd C:\Users\ChungkiLee\Documents\GitHub\webcam
import cv2
import numpy as np
import datetime
import sys
import time
import os
import shelve

# Create a VideoCapture object
cap = cv2.VideoCapture(0)
frameRate = 30.0
cap.set(cv2.CAP_PROP_FPS, frameRate)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 720) # 720, 1080
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 1280) # 1280, 1980
fps = int(cap.get(cv2.CAP_PROP_FPS))
print("The frame rate is: " + str(fps))

# Check if camera opened successfully
if (cap.isOpened() == False): 
    print("Unable to read camera feed")

# Define the codec and create VideoWriter object.The output is stored in 'outpy.avi' file.
tmp1 = str(datetime.datetime.now())
tmp1 = tmp1.replace(":", "_").replace(".", "_").replace(" ", "_").replace("-", "_")
tmp2 = 'C:/Users/ChungkiLee/Documents/GitHub/webcam/previewimgs/outpy'
tmp3 = tmp2 + tmp1[:19] +'.out'
filename = tmp3
del(tmp1, tmp2, tmp3)

imageEndingSeconds = 10;
imagePreparSeconds = 5;

cnt = 0; cntPre = 0;
a= 0.0; b=0.0; c= 0.0;
time_x = [] ; 
imgs = []; 
# 초기시간 
a = datetime.datetime.now()
print('time a is : ', a)
while(True):
    ret, frame = cap.read()
    frame = cv2.flip(frame, 1)
    
    if ret == True: 
        
        b = datetime.datetime.now()
        c = b - a
        if c.seconds > (imageEndingSeconds + imagePreparSeconds):    
            print('reason 0 problem is time', c.seconds, '  a?  ',a,'  b?  ',b)
            break
        # 이것은 image 데이터를 저장하는 곳
        if c.seconds > imagePreparSeconds:
            cv2.putText(frame, "image number: %s" % str(cnt-0), (10, 25), cv2.FONT_HERSHEY_PLAIN, 2, (100, 255, 100))
            cv2.putText(frame, "recording time(sec): %s" % str(c.seconds-imagePreparSeconds), (10, 50), cv2.FONT_HERSHEY_PLAIN, 2, (100, 255, 100))
            cnt += 1
            imgs.append(frame)
            time_x.append((c.seconds-imagePreparSeconds-1) + c.microseconds/1000000)
        else: 
            cv2.putText(frame, "Ready to the 90th frame: %s" % str(cntPre), (10, 25), cv2.FONT_HERSHEY_PLAIN, 2, (100, 255, 100))
            cv2.putText(frame, "recording time(sec): %s" % str(c.seconds), (10, 50), cv2.FONT_HERSHEY_PLAIN, 2, (100, 255, 100))
            cntPre += 1;
            
        # Display the resulting frame    
        cv2.imshow('frame',frame)

        # Press Q on keyboard to stop recording
        if cv2.waitKey(1) & 0xFF == ord('q'):
            print('reason 1')
            break
    # Break the loop
    else:
        print('reason 2')
        continue

# When everything done, release the video capture and video write objects
cap.release()
cv2.destroyAllWindows()

my_shelf = shelve.open(filename, 'n')
saving_files = ['imgs', 'time_x']
for key in saving_files:
    try:
        my_shelf[key] = globals()[key]
    except TypeError:
        print('error shelvinf : {0}'.format(key))
my_shelf.close()

print('-----------------------------------------------------------')
print('image stack is : ',np.array(imgs).shape)
print('time scale is : ',np.array(time_x).shape)
print('initial time is : ',np.array(time_x)[0], ' the end time is : ',np.array(time_x)[-1])
print('-----------------------------------------------------------')
print(time_x)
```
